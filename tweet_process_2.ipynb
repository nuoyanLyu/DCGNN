```python
# 添加时间信息、根据日期切片？
import datetime

def calcuTime(dateStr, r):
    date0 = datetime.datetime.strptime(str(dateStr)[:19], '%Y-%m-%d %H:%M:%S')
    date1 = datetime.datetime.strptime('2022-03-16 00:00:00', '%Y-%m-%d %H:%M:%S')
    delta = date1 - date0
    if r == 'days':
        return delta.days
    else:
        return delta.seconds

calcuTime('2021-03-15 20:14:35+00:00', 'days')
```

```python
dfLess['countDays'] = dfLess['created_at'].apply(lambda x: calcuTime(x, 'days'))
dfLess['countSeconds'] = dfLess['created_at'].apply(lambda x: calcuTime(x, 'seconds'))
dfLess
```
